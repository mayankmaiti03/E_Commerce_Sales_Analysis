```python
def find_outliers_iqr_multiple_columns(df, columns):
  outliers_dict = {}
  for column in columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)].index.tolist()
    outliers_dict[column] = outliers
  return outliers_dict

# Example usage:
file_path = c
df = pd.read_csv(file_path)

columns_to_check = ["Quantity","TotalValue","Price"]
outliers = find_outliers_iqr_multiple_columns(df, columns_to_check)

for column, indices in outliers.items():
  if indices:
    print(f"Outliers found in '{column}':")
    print(df.loc[indices, [column]])  # Print only the outlier values for the specific column
  else:
    print(f"No outliers found in '{column}'.")
```
